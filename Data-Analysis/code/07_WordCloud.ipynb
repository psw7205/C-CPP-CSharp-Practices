### 워드클라우드 설정

```python
wordcloud = WordCloud(max_font_size=300,
                     font_path='C:\Windows\Fonts\malgunbd.ttf',
                     stopwords=STOPWORDS,
                     background_color='#FFFFFF',
                     width=600, height=600).generate(text)
```

```python
wordcloud.words_
```

### 사이즈 설정

```python
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)

# 크기와 관련된 옵션 => 자주쓰임!

# 축과의 여백 설정 = 0
plt.tight_layout(pad=0)
# 축 안보이게 만들기
plt.axis('off')

# 저장
plt.savefig('../img/file.png', bbox_inches='tight')
```

## 배경이미지 Masked wordcloud

```python
import numpy as np
from PIL import Image

text = open('../data/wordcloud/alice.txt').read()
heart_mask = np.array(
    Image.open('../data/wordcloud/heart.jpg'))
stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color='white',
              max_words=2000, mask=heart_mask, stopwords=stopwords)
wc = wc.generate(text)
plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
